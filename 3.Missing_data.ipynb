```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() #is not a number
```

```python
df.isna().sum() #sum of NAN values
```

```python
df.isna().any() #used to check if at least one non-zero value exists
```

REMOVING MISSING DATA

```python
df
```

```python
df.dropna() #row wise drop
```

Threshold

```python
df.dropna(thresh=3) #row which has 3 true values
```

Filling the missing Data

```python
df
```

```python
df.fillna(0)
```

```python
values = {'A':0,'B':100,'C':300,'D':400} #column wise drop
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
